## Preprocessing + Feature Engineering

In the following cells we are going to be performing some feature engineering and data imputation on the rain_data data set.

```python
#create a new column 'fp_wins_round' that is True if 'fp_side' equals 'round_winner_side', otherwise False
rain_data['fp_wins_round'] = (rain_data['fp_side'] == rain_data['round_winner_side']).astype(bool)

rain_data[['fp_side', 'round_winner_side', 'fp_wins_round']].head()
```

```python
#create a new column 'fp_died' where it is true is if 'fp_death_loc' is not 'survived'
rain_data['fp_died'] = rain_data['fp_death_loc'].apply(lambda x: x != 'survived')       
rain_data[['fp_death_loc', 'fp_died']].head()
```

```python
#fill missing values in 'fp_kills' column with 0
rain_data['fp_kills'] = rain_data['fp_kills'].fillna(0) 
rain_data['fp_kills'].head()
#fill missing values in 'fp_kill_1_ location' column with 'no_kill'
rain_data['fp_kill_1_location'] = rain_data['fp_kill_1_ location'].fillna('no_kill')
#fill missing values in 'fp_kill_2_ location' column with 'no_kill'
rain_data['fp_kill_2_location'] = rain_data['fp_kill_2_ location'].fillna('no_kill')
#fill missing values in 'fp_kill_3_ location' column with 'no_kill'
rain_data['fp_kill_3_location'] = rain_data['fp_kill_3_ location'].fillna('no_kill')
#fill missing values in 'fp_kill_4_ location' column with 'no_kill'
rain_data['fp_kill_4_location'] = rain_data['fp_kill_4_ location'].fillna('no_kill')
#fill missing values in 'fp_kill_5_ location' column with 'no_kill'
rain_data['fp_kill_5_location'] = rain_data['fp_kill_5_ location'].fillna('no_kill')
```

```python
rain_data
```

```python
#fill missing values in each of the 'fp_wep_kill_x' columns with 'no_kill' 
rain_data['fp_wep_kill_1'] = rain_data['fp_wep_kill_1'].fillna('no_kill')
rain_data['fp_wep_kill_2'] = rain_data['fp_wep_kill_2'].fillna('no_kill')
rain_data['fp_wep_kill_3'] = rain_data['fp_wep_kill_3'].fillna('no_kill')
rain_data['fp_wep_kill_4'] = rain_data['fp_wep_kill_4'].fillna('no_kill')
rain_data['fp_wep_kill_5'] = rain_data['fp_wep_kill_5'].fillna('no_kill')
```

```python
rain_data = rain_data.drop(columns=['fp_kill_1_ location', 'fp_kill_2_ location', 'fp_kill_3_ location', 'fp_kill_4_ location', 'fp_kill_5_ location'])
```

```python
#change datatype of opening_kill to boolean where 1 is True and 0 is False
rain_data['opening_kill'] = rain_data['opening_kill'].astype(bool)
```

```python
rain_data.describe()
```

```python
rain_data.columns
```

```python
#creating a new column FP_Kills_Last_3_Rounds, Sum of fp_kills from the last 3 rounds (N-1 to N-3).Captures the player's short-term form and confidence. A player on a hot streak might be more aggressive. but only include rounds that are from the same match id   
rain_data['FP_Kills_Last_3_Rounds'] = 0
for i in range(len(rain_data)):
    match_id = rain_data.loc[i, 'match_id']
    round_num = rain_data.loc[i, 'round_num']
    last_3_rounds = rain_data[(rain_data['match_id'] == match_id) & (rain_data['round_num'] < round_num) & (rain_data['round_num'] >= round_num - 3)]
    rain_data.loc[i, 'FP_Kills_Last_3_Rounds'] = last_3_rounds['fp_kills'].sum()
rain_data[['match_id', 'round_num', 'fp_kills', 'FP_Kills_Last_3_Rounds']].head(10)
```

```python
#creating a new column fp_team_win_streak, counts the number of consecutive rounds the player's team has won up to the current round within the same match
rain_data['fp_team_win_streak'] = 0
for i in range(len(rain_data)):
    match_id = rain_data.loc[i, 'match_id']
    round_num = rain_data.loc[i, 'round_num']
    fp_side = rain_data.loc[i, 'fp_side']
    
    win_streak = 0
    for j in range(round_num - 1, 0, -1):
        previous_round = rain_data[(rain_data['match_id'] == match_id) & (rain_data['round_num'] == j)]
        if not previous_round.empty:
            if previous_round.iloc[0]['round_winner_side'] == fp_side:
                win_streak += 1
            else:
                break
    rain_data.loc[i, 'fp_team_win_streak'] = win_streak
rain_data[['match_id', 'round_num', 'fp_side', 'round_winner_side', 'fp_team_win_streak']].head(10)
```

```python
#creating a new column opp_team_win_streak, counts the number of consecutive rounds the opposing team has won up to the current round within the same match
rain_data['opp_team_win_streak'] = 0   
for i in range(len(rain_data)):
    match_id = rain_data.loc[i, 'match_id']
    round_num = rain_data.loc[i, 'round_num']
    fp_side = rain_data.loc[i, 'fp_side']
    opp_side = 'ct' if fp_side == 't' else 't'
    
    win_streak = 0
    for j in range(round_num - 1, 0, -1):
        previous_round = rain_data[(rain_data['match_id'] == match_id) & (rain_data['round_num'] == j)]
        if not previous_round.empty:
            if previous_round.iloc[0]['round_winner_side'] == opp_side:
                win_streak += 1
            else:
                break
    rain_data.loc[i, 'opp_team_win_streak'] = win_streak
rain_data[['match_id', 'round_num', 'fp_side', 'round_winner_side', 'opp_team_win_streak']].head(10)
```

### Consistency and Momentum

The first thing I want to analyze is rains consistency and momentum. How reliable is he when the odds are against him? Is he a an unchanging rock that performs no matter the circumstances? Or does he crumble under the pressure? How well does he carry his momentum between rounds? Is he able to snowball when he enters the flow state and can't be stopped? Let's find out.

#### Consistency

We are going to first look at whether or not rain's performance is personally affected when his opponent has momentum, is he able to step up or crumble under pressure?

```python
streak_performance = rain_data.groupby('opp_team_win_streak')['fp_kills'].mean().reset_index()
streak_performance.columns = ['Opponent_Streak_Length', 'FP_Average_Kills']

print("--- FP Average Kills vs. Opponent Win Streak ---")
print(streak_performance)

# --- 3. Binning for Clearer Analysis (Optional but recommended) ---
# If you have a long tail of streak lengths (e.g., streaks of 7, 8, 9...),
# it's best to group them into an '4+ Streak' category for statistical robustness.
bins = [-1, 0, 1, 2, 3, rain_data['opp_team_win_streak'].max()]
labels = ['Streak_0', 'Streak_1', 'Streak_2', 'Streak_3', 'Streak_4+']

rain_data['Streak_Group'] = pd.cut(
    rain_data['opp_team_win_streak'],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)

binned_performance = rain_data.groupby('Streak_Group', observed=False)['fp_kills'].agg(['mean', 'count']).reset_index()
binned_performance.columns = ['Opponent_Streak_Group', 'FP_Average_Kills', 'Round_Count']

print("\n--- FP Average Kills vs. Binned Opponent Win Streak ---")
print(binned_performance)
```

```python
plt.style.use('ggplot')

fig, ax1 = plt.subplots(figsize=(10, 6))

colors = plt.cm.plasma(np.linspace(0.2, 0.8, len(binned_performance)))

bars = ax1.bar(
    binned_performance['Opponent_Streak_Group'],
    binned_performance['FP_Average_Kills'],
    color=colors,
    alpha=0.8,
    label='Average Kills (per Round)'
)

for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval + 0.05,
             f'{yval:.2f}', ha='center', va='bottom', fontsize=10, color='black')

ax1.axhline(
    0.777, #0.777 is rain's average kills per round
    color='green',
    linestyle='--',
    linewidth=2,
    label=f'FP Overall Average (0.777)' 
)

ax1.set_ylim(0, binned_performance['FP_Average_Kills'].max() * 1.3)
ax1.set_xlabel('Opponent Win Streak (Pressure Level)', fontsize=12)
ax1.set_ylabel('FP Average Kills per Round', color='darkblue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='darkblue')
ax1.grid(axis='y', linestyle='--', alpha=0.6)

ax2 = ax1.twinx()
ax2.plot(
    binned_performance['Opponent_Streak_Group'],
    binned_performance['Round_Count'],
    color='red',
    marker='o',
    linestyle='--',
    linewidth=1,
    label='Total Rounds in Group'
)

ax2.set_ylabel('Total Rounds in Group (Count)', color='red', fontsize=12)
ax2.tick_params(axis='y', labelcolor='red')

for i, count in enumerate(binned_performance['Round_Count']):
    ax2.text(i, count + 5, f'N={count}', ha='center', color='red', fontsize=9)


plt.title(
    "Focus Player (FP) Performance vs. Opponent Momentum",
    fontsize=14,
    fontweight='bold'
)
fig.tight_layout() 
plt.show()

print("The colored bars show the FP's average kills against each opponent streak group (0,1,2,3,4+), and the red line/dots show the number of rounds used for that average (sample size).")
print("And the green dashed line indicates the FP's overall average kills per round (0.777). This visualization helps assess how the FP performs under varying levels of opponent momentum pressure.")
```

So as we see above from the visualization rain is able to keep his perfomance, kill-wise, consistent as long as the opponent's round streak remains under 3 rounds. As we get into the deeper opponent round streaks 3 and above we see a steep drop off in consistency, which indicates reliability until the opponents have won 3 rounds in a row in which he seems to lose momentum and potentially confidence to get kills.

#### Momentum

Now that we looked at how rain deals with his opposition's momentum, let's see how he deals with his own? Is he a momentum based player that can go on a hot streak and enter a "flow-state"? Or does he have a stable unmoving performance (kill-wise) even when the kills start rolling in?

```python
correlation_data = rain_data[['fp_kills', 'FP_Kills_Last_3_Rounds']]

correlation_matrix = correlation_data.corr()

plt.figure(figsize=(7, 6))

sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt='.4f',
    cmap='coolwarm',
    cbar=True,
    linewidths=.5,
    linecolor='black',
    square=True
)

plt.title(
    'Correlation Between Current Performance and Previous Momentum',
    fontsize=14,
    fontweight='bold'
)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()

momentum_correlation = correlation_matrix.loc['fp_kills', 'FP_Kills_Last_3_Rounds']

print("\n--- Correlation Analysis ---")
print(f"Correlation (r) between current kills and kills in last 3 rounds: {momentum_correlation:.4f}")

if momentum_correlation > 0.3:
    print("Interpretation: There is a POSITIVE and MODERATE-TO-STRONG correlation.")
    print("This suggests the player is MOMENTUM-DEPENDENT: high kills in previous rounds lead to high kills in the current round.")
elif momentum_correlation < -0.3:
    print("Interpretation: There is a NEGATIVE and MODERATE-TO-STRONG correlation.")
    print("This suggests a REVERSION TO THE MEAN: high kills previously might indicate they are due for a lower-kill round (or vice-versa).")
else:
    print("Interpretation: The correlation is WEAK or negligible.")
    print("This suggests the player's performance in any given round is largely INDEPENDENT of their immediate historical performance.")
```

As the statement above concluded, rain appears to be able to perform and get kills completely independent of his previous rounds performance indicating that his kills come from consistency and not due to him "heating up".

### Economic and Contextual Performance

In this section we will focus on rain's performance changes dependent on his team's and opponent's money situation. Is he able to perform as well under poor economic conditions as he does under ideal ones? Or is he only a sunshine shooter?

#### Low Buy/Full Buy Kill Efficiency

Here we will analyze the context of rain's kills, whether he got most of his kills against low buys, half buys, full buys or an even distribution of them.

```python
# Group the data by the opponent's round type and calculate the mean kills
efficiency_data = rain_data.groupby('opp_round_type')['fp_kills'].agg(['mean', 'count']).reset_index()
efficiency_data.columns = ['Opponent_Round_Type', 'FP_Average_Kills', 'Round_Count']

# Define the order of the round types (from strongest opposition to weakest)
round_order = ['full', 'half', 'low', 'pistol']
efficiency_data['Opponent_Round_Type'] = pd.Categorical(
    efficiency_data['Opponent_Round_Type'],
    categories=round_order,
    ordered=True
)
efficiency_data = efficiency_data.sort_values('Opponent_Round_Type')


# --- 3. Visualization: Bar Chart of KPR by Opponent Round Type ---

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(9, 6))

# Use a colormap, making the "full" buy rounds visually distinct if possible
colors = sns.color_palette("viridis", len(efficiency_data))

# Create the Bar Chart
bars = ax.bar(
    efficiency_data['Opponent_Round_Type'],
    efficiency_data['FP_Average_Kills'],
    color=colors,
    alpha=0.9
)

# Add value labels (KPR) on top of the bars and round count (N) inside
for i, bar in enumerate(bars):
    yval = bar.get_height()
    N = efficiency_data['Round_Count'].iloc[i]

    # KPR Label (on top)
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.1,
            f'{yval:.2f}', ha='center', va='bottom', fontsize=11, color='black', fontweight='bold')

    # Round Count Label (inside)
    ax.text(bar.get_x() + bar.get_width()/2, 0.1,
            f'N={N}', ha='center', va='bottom', fontsize=9, color='white')


# --- 4. Final Touches and Interpretation ---

# Extract average kills for Full and Low for the key comparison
full_kpr = efficiency_data[efficiency_data['Opponent_Round_Type'] == 'full']['FP_Average_Kills'].iloc[0]
low_kpr = efficiency_data[efficiency_data['Opponent_Round_Type'] == 'low']['FP_Average_Kills'].iloc[0]
kpr_difference = full_kpr - low_kpr

ax.set_ylim(0, efficiency_data['FP_Average_Kills'].max() * 1.3)
ax.set_xlabel('Opponent Buy Type (Strength)', fontsize=12)
ax.set_ylabel('FP Average Kills per Round (KPR)', fontsize=12)

# Highlight the most important comparison
plt.title(
    "Focus Player Kill Efficiency: KPR vs. Opponent Buy Type",
    fontsize=14,
    fontweight='bold'
)
plt.suptitle(
    f"Difference in KPR (Full vs. Low): {kpr_difference:.2f}",
    fontsize=10,
    color='gray',
    y=0.92
)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

print(f"Average KPR against 'full' buy opponents: {full_kpr:.2f}")
print(f"Average KPR against 'low' buy opponents: {low_kpr:.2f}")
```

As the graph above shows, the difference between rains average kills against full buy opponents and low buy opponents is only 0.05 indicating that rain is almost equally as effective against a fully equipped opponent as he is against a poorly equipped one. 

#### Investment vs. Impact

Now that we know rain is effective regardless of the opponents economic situation, lets see how efficient he is by creating the metric "kills-per-$1000". This will answer the question, does rain have the ability to get kills even with subpar equipment? or does he only perform when fully stacked?

```python
rain_data['equip_value_k'] = rain_data['fp_equip_value'] / 1000


rain_data['kpr_per_k_spent'] = rain_data.apply(
    lambda row: row['fp_kills'] / row['equip_value_k'] if row['equip_value_k'] > 0 else 0,
    axis=1
)


def classify_fp_buy(value):
    if value < 1800:
        return 'Eco Buy (Low Investment)'
    elif value <= 3500:
        return 'Half Buy (Mid Investment)'
    else:
        return 'Full Buy (High Investment)'

rain_data['fp_buy_type'] = rain_data['fp_equip_value'].apply(classify_fp_buy)


analysis_summary = rain_data.groupby('fp_buy_type')['kpr_per_k_spent'].agg(['mean', 'std', 'count']).reset_index()
analysis_summary.columns = ['FP_Buy_Type', 'Avg_KPR_Per_K_Spent', 'Std_Dev', 'Round_Count']


buy_order = ['Eco Buy (Low Investment)', 'Half Buy (Mid Investment)', 'Full Buy (High Investment)']
analysis_summary['FP_Buy_Type'] = pd.Categorical(
    analysis_summary['FP_Buy_Type'],
    categories=buy_order,
    ordered=True
)
analysis_summary = analysis_summary.sort_values('FP_Buy_Type')



plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(
    analysis_summary['FP_Buy_Type'],
    analysis_summary['Avg_KPR_Per_K_Spent'],
    color=['#1f77b4', '#ff7f0e', '#2ca02c'], 
    alpha=0.9
)


for i, bar in enumerate(bars):
    yval = bar.get_height()
    N = analysis_summary['Round_Count'].iloc[i]
    
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.02,
            f'{yval:.3f}', ha='center', va='bottom', fontsize=11, color='black', fontweight='bold')
    
    ax.text(bar.get_x() + bar.get_width()/2, 0.05,
            f'N={N}', ha='center', va='bottom', fontsize=9, color='white')


ax.set_ylim(0, analysis_summary['Avg_KPR_Per_K_Spent'].max() * 1.2)
ax.set_xlabel("Focus Player's Investment Level", fontsize=12)
ax.set_ylabel('Average Kills Per $1000 Spent (KPR/$k)', fontsize=12)

plt.title(
    "Focus Player's Economic Impact: KPR/$k by Buy Type",
    fontsize=15,
    fontweight='bold'
)
plt.suptitle(
    "A higher KPR/$k means more efficient kill-getting for less money.",
    fontsize=10,
    color='gray',
    y=0.92
)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

print("\n--- Investment vs. Impact Analysis Summary ---")
print("\nInterpretation:")
print(f"The player's highest economic efficiency (best KPR/$k) is during the '{analysis_summary.iloc[analysis_summary['Avg_KPR_Per_K_Spent'].argmax()]['FP_Buy_Type']}' rounds.")
```

Then above graph shows that rain's kill per $1000 is extremely high during his eco/low buy rounds, alluding to him being an economic machine, with the ability to provide huge impact even with cheap gear.

```python
POSSIBLE_LOCATIONS = [
    'cat', 'b_site', 'b_ramp', 'cave', 'a_site', 'mid', 'donut'
]

NO_LOCATION_STR = 'no_kill'

OTHERS_THRESHOLD = 0.03

kill_cols = [col for col in rain_data.columns if 'fp_kill' in col]

rain_data[kill_cols] = rain_data[kill_cols].astype(str).fillna(NO_LOCATION_STR) 

melted_df = rain_data.melt(
    id_vars=['fp_side'], 
    value_vars=kill_cols, 
    value_name='Kill_Location'
)

all_kill_locations = melted_df[melted_df['Kill_Location'].isin(POSSIBLE_LOCATIONS)].copy() 

all_kill_counts = all_kill_locations['Kill_Location'].value_counts()
total_valid_kills = all_kill_counts.sum()

kill_percentages = all_kill_counts / total_valid_kills
small_locations = kill_percentages[kill_percentages < OTHERS_THRESHOLD]

if not small_locations.empty:
    others_count = all_kill_counts.loc[small_locations.index].sum()
    top_counts = all_kill_counts.drop(small_locations.index)
    if others_count > 0:
        top_counts['Others'] = others_count
else:
    top_counts = all_kill_counts.copy()

top_counts = top_counts.sort_values(ascending=False)
labels = top_counts.index.tolist()
sizes = top_counts.values.tolist()


plt.style.use('seaborn-v0_8-whitegrid')

kill_counts_iter = iter(sizes)
def kill_count_and_pct(pct):
    absolute_count = next(kill_counts_iter)
    
    return f'{absolute_count} ({pct:.1f}%)'

fig, ax = plt.subplots(figsize=(14, 14))

colors = plt.cm.Set3.colors 

wedges, texts, autotexts = ax.pie(
    sizes, 
    labels=labels, 
    autopct=kill_count_and_pct, 
    startangle=90,     
    colors=colors,
    wedgeprops={"edgecolor": "black", 'linewidth': 1, 'antialiased': True}
)

for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')
for text in texts:
    text.set_fontsize(13)

ax.axis('equal')  

ax.set_title(
    f'Overall Kill Location Distribution (Threshold for "Others": {OTHERS_THRESHOLD*100:.0f}%)', 
    fontsize=20, 
    fontweight='bold', 
    pad=20
)

plt.show()
```

#### Kill Significance

This next analysis is on rains kill significance, we are going to look at his average kills in wins minus the average kills in losses to see how important him getting kills is to winning the round.

```python
try:
    rain_data['fp_kills'] = pd.to_numeric(rain_data['fp_kills'], errors='coerce').fillna(0).astype(int)
except Exception as e:
    print(f"Warning: Could not convert 'fp_kills' to integer. Error: {e}")
    # Proceeding with float conversion if int fails
    rain_data['fp_kills'] = pd.to_numeric(rain_data['fp_kills'], errors='coerce')
total_rounds = rain_data.shape[0]

rain_data['Round_Result'] = np.where(
    rain_data['fp_wins_round'] == True, 
    'Win', 
    'Loss'
)


kill_value_summary = rain_data.groupby('Round_Result')['fp_kills'].agg(['count', 'sum', 'mean', 'std']).reset_index()
kill_value_summary = kill_value_summary.rename(columns={
    'count': 'Rounds',
    'sum': 'Total Kills',
    'mean': 'Avg Kills per Round',
    'std': 'Std Dev of Kills'
})


avg_kills_win = kill_value_summary[kill_value_summary['Round_Result'] == 'Win']['Avg Kills per Round'].iloc[0]
avg_kills_loss = kill_value_summary[kill_value_summary['Round_Result'] == 'Loss']['Avg Kills per Round'].iloc[0]


kill_value_difference = avg_kills_win - avg_kills_loss



print("="*60)
print(f"KILL VALUE & SIGNIFICANCE ANALYSIS (Total Rounds: {total_rounds})")
print("="*60)


print("\n" + "-"*60)
print(f"AVERAGE KILLS IN WINS: {avg_kills_win:.2f}")
print(f"AVERAGE KILLS IN LOSSES: {avg_kills_loss:.2f}")
print(f"KILL VALUE DIFFERENCE (Win Avg - Loss Avg): {kill_value_difference:.2f}")
print("-"*60)


print("\n--- Interpretation: Kill Significance ---")

if kill_value_difference > 0.1:
    print(f"The Focus Player averages {kill_value_difference:.2f} **more kills** in rounds their team WINS.")
    print("This indicates a **High Kill Value**. Their kills are highly significant, meaning they are likely:")
    print("1. Securing crucial **entry kills** or mid-round kills that directly create a team advantage.")
    print("2. The source of the necessary firepower to close out competitive rounds.")
elif kill_value_difference < -0.1:
    print(f"The Focus Player averages {abs(kill_value_difference):.2f} **fewer kills** in rounds their team WINS.")
    print("This indicates a **Low Kill Value**.")
    print("1. They might be getting many 'junk kills' or exit frags after the round is already lost.")
    print("2. While their output is high, it is not strongly correlated with team success.")
else:
    print("The average kills in Wins and Losses are nearly identical.")
    print("This suggests their kill output is **highly consistent** but doesn't necessarily spike during successful rounds.")
```
